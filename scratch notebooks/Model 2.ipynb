```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('raw data/msft.csv')
```

```python
df.columns
```

```python
def to_log_rtn(df, col='5. adjusted close', shift=1):
   
    """
    df is a pandas dataframe
    col is the name of the column corresponding to the price
    shift is the number of days for the return (5 = weekly, 20 = monthly, etc...) in business days
    """
    
    return np.log(df[col]/df[col].shift(shift))
```

```python
def to_log_HL(df, high, low, shift=1):
    
    """
    df is a pandas dataframe
    col is the name of the column corresponding to the price
    shift is the number of days for the return (5 = weekly, 20 = monthly, etc...) in business days
    """
    
    return np.log(df[high].rolling(window=shift).max()/df[low].rolling(window=shift).min())  
```

```python
daily_return = np.log(df['5. adjusted close']/df['5. adjusted close'].shift(1))
daily_return.tail()
```

```python
weekly_return = np.log(df['5. adjusted close']/df['5. adjusted close'].shift(5))
weekly_return.tail()
```

```python
dailyHL = np.log(df['2. high']/df['3. low'])
dailyHL.tail()
```

```python
weeklyHL = np.log(df['2. high'].rolling(window=5).max()/df['3. low'].rolling(window=5).min())
weeklyHL.tail()
```

```python
data = pd.concat([daily_return, weekly_return, dailyHL, weeklyHL], axis=1)
data.columns = ['daily_return', 'weekly_return', 'daliyHL', 'weeklyHL']
data.dropna(inplace=True)
```

```python
def toTimeSeries(df, columns, trail, lead):
    """
    custom function to create supervised dataset with trailing & leading data
    """
    df = df[columns]
    n_columns = len(columns)
    cols = []
    names = []
    #create trail
    for i in range(trail, 0, -1):
        cols.append(df.shift(i))
        names += [(columns[j] + '(t - %d)' % (i)) for j in range(n_columns)]
    
    #create lead
    for i in range(0, lead+1):
        cols.append(df.shift(-i))
        names += [(columns[j] + '(t + %d)' % (i)) for j in range(n_columns)]
        
    #aggregate lead & trail
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    agg.dropna(inplace=True)
    
    return agg
```

```python
n_trail = 5
n_lead = 5
columns = ['daily_return', 'weekly_return', 'daliyHL', 'weeklyHL']
reframed = toTimeSeries(data, columns, n_trail, n_lead)
```

```python
split = 0.75  # portion of data used for training

values = reframed.values

mask = np.random.rand(len(values)) < split

train = values[mask]
test = values[~mask]
```

```python
n_columns = len(columns)
n_features = (n_lead - 1) *  len(columns)
n_periods = n_trail + 1
n_obs = n_periods * n_columns
```

```python
X_train, y_train = train[:, 0:n_obs], train[:, -n_features:]
X_test, y_test = test[:, 0:n_obs], test[:, -n_features:]
```

```python
X_train = X_train.reshape((X_train.shape[0], n_periods, n_columns))
X_test = X_test.reshape((X_test.shape[0], n_periods, n_columns))
```

```python
new = []

for i, train in enumerate(y_train):
    new.append(y_train[i][-n_columns:].tolist())
    
y_train = np.array(new)

new = []

for i, test in enumerate(y_test):
    new.append(y_test[i][-n_columns:].tolist())
    
y_test = np.array(new)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1:])))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(len(columns)))
model.compile(loss='mse', optimizer='adam')
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=72,
                     validation_data=(X_test, y_test), shuffle=False)
```

```python
plt.figure(figsize=(15,6))
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()

plt.ylabel('error')
plt.xlabel('epoch')
plt.show()
```

```python
prd = model.predict(X_test)
```

```python
results = pd.DataFrame(prd)
results.columns = columns
real = pd.DataFrame(y_test, columns=columns)
real.columns += '_y_test'

results = pd.concat([results, real], axis=1)

plt.figure(figsize=(15,6))
plt.plot(results.daily_return, label='predicted')
plt.plot(results.daily_return_y_test, label='true')
plt.legend()
plt.show()
```

```python
results.columns
```

```python
results['correct_daily_return_up'] = (results.daily_return*results.daily_return_y_test).apply(lambda x: 1 if x > 0 else 0)
results['correct_weekly_return_up'] = (results.weekly_return*results.weekly_return_y_test).apply(lambda x: 1 if x > 0 else 0)
```

```python
print('correctly predicted daily return increase:', results.correct_daily_return_up.sum()/len(results))
print('correctly predicted weekly return increase:', results.correct_weekly_return_up.sum()/len(results))
```

```python
def get_stock_data(symbol, start_date='2010-01-01', end_date='2018-12-31'):
    
    """
    function to pull adjusted stock prices from the alphavantage API.
    
    symbol is a string representing a stock symbol, e.g. 'AAPL'
    
    start_date is the start date of the time series, defaults to '1/1/2010'
    
    end_date is the end date of the time series, defaults to '12/31/2018'
    
    start_date & end_date must be in string format
    """
    
    url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=' + symbol + '&outputsize=full&apikey=' + api_key
        
    r = requests.get(url)
    
    df = r.json()
    
    df = pd.DataFrame(df['Time Series (Daily)']).T
    
    df = df[(df.index > start_date) & (df.index < end_date)]
 
    return df
```
